```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(style="whitegrid")
```

```python
# Choose any one dataset from seaborn's built-in datasets (e.g., car_crashes, tips,
# taxis, penguins, flights, diamonds).
data = sns.load_dataset("tips")
```

```python
# Perform basic data exploration and summarization.
print(data.head())
print("\n\n",data.describe())
```

```python
print(data.info())
print("\n\n",data.isnull().sum())
```

```python
print(data.describe(include='all'))
```

```python
print("\n\n",data.nunique())
```

```python
print("\n\n",data['time'].value_counts())
```

```python
print("\n\n",data['day'].value_counts())
```

```python
# Univariate analysis:
sns.histplot(data['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x="day")
plt.title("Count of Records by Day")
plt.show()
```

```python
# Bivariate analysis:
plt.figure(figsize=(6, 4))
sns.scatterplot(data=data, x="total_bill", y="tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x="sex", y="tip")
plt.title("Tip Distribution by Gender")
plt.show()
```

```python
# Multivariate analysis:
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="total_bill", y="tip", hue="day", style="smoker")
plt.title("Total Bill vs Tip by Day and Smoker Status")
plt.legend(title='Day and Smoker Status')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=data, x="total_bill", y="tip", hue="time", style="sex")
plt.title("Total Bill vs Tip by Time and Gender")
plt.show()
```


### Insights

1. Higher bills tend to result in higher tips
3. Males and females tip similar but males are slightly more.
2. Dinner time tends to involve higher bills and tips compared to lunch.
4. Difference between smoker and non-somker is not very significant.
